```python
# to drop the null values
df.dropna(inplace=True)
```

```python
ax = sns.countplot(x='Gender', data=df)

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 0.1, int(height), ha="center")
plt.show()
```

```python
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
plt.show()
```

# from above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than males

# Age

```python
ax = sns.countplot(x='Age Group', hue='Gender', data=df) 

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 0.1, int(height), ha="center")
plt.show()
```

```python
sales_gen = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_gen)
plt.show()
```

# from above graphes we can see that most of the buyers are of age group between 26-35 yrs females

# states

```python
# total numbers of orders from top 10 states
sales_state = df.groupby('State', as_index=False)['Orders'] \
                .sum() \
                .sort_values(by='Orders', ascending=False) \
                .head(10)

sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
plt.title("Top 10 States by Orders")
plt.show()
```

```python
# total amount/sales from top 10 states
sales_state = df.groupby('State', as_index=False)['Amount'] \
                .sum() \
                .sort_values(by='Amount', ascending=False) \
                .head(10)

sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(x='State', y='Amount', data=sales_state)
plt.show()
```

# from above graphes we can see that most of the orders and total sales/amount are from uttarpradesh, maharashtra and karnataka respectively
